```python
X1 = 4
X2 = 3
X3 = 1
X4 = 2
X = np.array([X1,X2,X3,X4])
```

```python
w1 = X1/(X1+X2+X3+X4)
w2 = X2/(X1+X2+X3+X4)
w3 = X3/(X1+X2+X3+X4)
w4 = X4/(X1+X2+X3+X4)
W = np.array([w1,w2,w3,w4])
```

```python
Cov = np.array([[1,0.489,0.496,-0.062],[0.489,1,0.918,0.201],[0.496,0.918,1,0.211],[-0.062,0.201,0.211,1]])
```

```python
Var = np.matmul(np.matmul(W.T,Cov),W)
vol = np.sqrt(Var)
```
